# AI-Based Financial Time Series Forecasting using LSTM

# Data Collection

## Objectives

- Download stock market data
- Explore the dataset
- Visualize stock prices
- Save the dataset for preprocessing

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Loaded Successfully!")
```

## Choose Stock

```python
ticker = "AAPL"
```

## Download Data

```python
stock_data = yf.download(
    ticker,
    start="2015-01-01",
    end="2025-01-01"
)

stock_data.head()
```

## Dataset Information

```python
stock_data.info()
```

## Missing Values

```python
stock_data.isnull().sum()
```

## Statistical Summary

```python
stock_data.describe()
```

## Closing Price

```python
plt.figure(figsize=(12,6))

plt.plot(
    stock_data["Close"],
    label="Closing Price"
)

plt.title("Apple Stock Closing Price")

plt.xlabel("Date")

plt.ylabel("Price (USD)")

plt.legend()

plt.grid(True)

plt.show()
```

## Trading Volume

```python
plt.figure(figsize=(12,6))

plt.plot(
    stock_data["Volume"],
    color="orange"
)

plt.title("Trading Volume")

plt.xlabel("Date")

plt.ylabel("Volume")

plt.grid(True)

plt.show()
```

## Moving Average - 50 Days

```python
stock_data["MA50"] = stock_data["Close"].rolling(50).mean()

plt.figure(figsize=(12,6))

plt.plot(
    stock_data["Close"],
    label="Close"
)

plt.plot(
    stock_data["MA50"],
    label="50-Day Moving Average"
)

plt.title("Closing Price vs 50-Day Moving Average")

plt.legend()

plt.grid(True)

plt.show()
```

## Moving Average - 200 Days

```python
stock_data["MA200"] = stock_data["Close"].rolling(200).mean()

plt.figure(figsize=(12,6))

plt.plot(
    stock_data["Close"],
    label="Close"
)

plt.plot(
    stock_data["MA200"],
    label="200-Day Moving Average"
)

plt.title("Closing Price vs 200-Day Moving Average")

plt.legend()

plt.grid(True)

plt.show()
```

## Save Dataset

```python
stock_data.to_csv("../data/stock_data.csv")

print("Dataset Saved Successfully!")
```
