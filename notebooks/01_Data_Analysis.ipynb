# Data Analysis

## Objectives

- Load Apple stock dataset
- Explore the dataset
- Check missing values
- Visualize stock prices
- Analyze market trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries Loaded Successfully!")
```

## Load Dataset

```python
data = pd.read_csv("../data/AAPL.csv")

data.head()
```

## Dataset Information

```python
data.info()
```

## Missing Values

```python
data.isnull().sum()
```

## Statistical Summary

```python
data.describe()
```

## Convert Data

```python
data["Date"] = pd.to_datetime(data["Date"])

data = data.sort_values("Date")

data.head()
```

## Closing Price Graph

```python
plt.figure(figsize=(14,6))

plt.plot(
    data["Date"],
    data["Close"],
    color="blue"
)

plt.title("Apple Closing Stock Price")

plt.xlabel("Date")

plt.ylabel("Price (USD)")

plt.grid(True)

plt.show()
```

## Trading Volume

```python
plt.figure(figsize=(14,6))

plt.plot(
    data["Date"],
    data["Volume"],
    color="orange"
)

plt.title("Trading Volume")

plt.xlabel("Date")

plt.ylabel("Volume")

plt.grid(True)

plt.show()
```

## 50-Day Moving Average

```python
data["MA50"] = data["Close"].rolling(50).mean()

plt.figure(figsize=(14,6))

plt.plot(
    data["Date"],
    data["Close"],
    label="Close Price"
)

plt.plot(
    data["Date"],
    data["MA50"],
    label="50-Day MA"
)

plt.title("Closing Price vs 50-Day Moving Average")

plt.legend()

plt.grid(True)

plt.show()
```

## 200-Day Moving Average

```python
data["MA200"] = data["Close"].rolling(200).mean()

plt.figure(figsize=(14,6))

plt.plot(
    data["Date"],
    data["Close"],
    label="Close Price"
)

plt.plot(
    data["Date"],
    data["MA200"],
    label="200-Day MA"
)

plt.title("Closing Price vs 200-Day Moving Average")

plt.legend()

plt.grid(True)

plt.show()
```

## Highest Closing Price

```python
highest = data.loc[data["Close"].idxmax()]

print("Highest Closing Price")
print(highest)
```

## Lowest Closing Price

```python
lowest = data.loc[data["Close"].idxmin()]

print("Lowest Closing Price")
print(lowest)
```

## Save Clean Dataset

```python
data.to_csv(
    "../data/clean_stock_data.csv",
    index=False
)

print("Clean dataset saved successfully!")
```
